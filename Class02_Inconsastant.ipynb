```python
import pandas as pd
import numpy as np
```

```python
data = {
    'date': ['2021-12-01','01-12-2022','2022/12/01','12-01-2021'],
    'country': ['USA', 'U.S.A', 'Amrica','United States'],
    'name': ['John Doe', 'Jonh Doe', 'Jane Doe','Jane Doe'],
    'sales_2023': [100, 200,None, 200],
    'sales_2024': [None,150, 300, 150]
}

df = pd.DataFrame(data)
df.head()
```

```python
# df['date'] = pd.to_datetime(df['date'], errors='coerce')
# df['date'] = df['date'].dt.strftime('%y-%m-%d')
frmt  = ['%Y-%m-%d']
def parse_date(s):
    # Replace '/' with '-' for consistent formatting
    s = s.replace('/', '-')
    # Try parsing with different formats
    for fmt in ['%Y-%m-%d', '%d-%m-%Y', '%m-%d-%Y']:
        try:
            return pd.to_datetime(s, format=frmt)
        except ValueError:
            continue
    return pd.NaT  # Return NaT for unparseable formats

# Apply the parsing function to the 'date' column
df['date'] = df['date'].apply(parse_date)

# Format to 'yy-mm-dd'
df['date'] = df['date'].dt.strftime('%y-%m-%d')

# Display the updated DataFrame
df.head()
```

```python
country_maping = {'USA':'United States','U.S.A':'United States','Amrica':'United States',}
df['country'] = df['country'].replace(country_maping)
df.head()
```

```python
df['name'] = df['name'].replace({'Jonh Doe':'John Doe','Jane Doe':'John Doe'})
df
```

```python
df = df.drop(df[df['sales_2024'] <= df['sales_2023']].index)
df
```
