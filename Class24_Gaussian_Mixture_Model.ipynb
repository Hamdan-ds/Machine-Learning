## Gaussian Mixture Model
A Gaussian Mixture Model (GMM) is a probabilistic model that assumes data is generated from a combination (mixture) of several Gaussian distributions. Each Gaussian has its own mean (center) and variance (spread), and the model tries to figure out how many Gaussians are needed and how they overlap.

- Mixture of Gaussians: Instead of one bell curve, the data is explained by multiple bell curves added together.
- Soft Clustering: Unlike K-Means (which assigns each point to one cluster), GMM gives each point a probability of belonging to each cluster.

## Parameters:
- Mean (μ): the center of each Gaussian.
- Variance (σ²): how wide the Gaussian is.
- Weight: how much each Gaussian contributes to the overall mixture.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
from sklearn import datasets
```

```python
iris = datasets.load_iris()
x = iris.data[:,:2]

gmm = GaussianMixture(n_components=3)
gmm.fit(x)
labels = gmm.predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=labels,cmap='viridis')
plt.show()
```

## Log Likelihood
- Maximum Likelihood Estimation (MLE): We adjust the mean and variance (or covariance matrix) until the log-likelihood is maximized. That gives us the “best fit” Gaussian parameters for the data.
- In machine learning, this is the foundation for algorithms like Gaussian Mixture Models (GMMs), clustering, and probabilistic models.

### Think of log-likelihood as a scorecard:
- High score → your Gaussian parameters explain the data well.
- Low score → your parameters don’t match the data.

```python
log_likelihood = gmm.score(x)
print('Log-Likelihood',log_likelihood)
```

## AIC (Akaike Information Criterition) :
  AIC: Focuses on finding the model that best predicts future data. It balances fit and complexity but is more forgiving about adding extra parameters.
## BIC (Bayesian Information Criterition) : 
  BIC: Similar idea, but it penalizes complexity more strongly, especially when the dataset is large. It leans toward simpler models.

### How to Interpret
- Both give you a score for each model.
- Lower score = better model (relative to the others you’re comparing).
- You don’t interpret the absolute value of AIC/BIC; you only compare them across models.


```python
aic = gmm.aic(x)
bic = gmm.bic(x)

print('AIC', aic)  # Akaike Information Criterition
print('BIC', bic) # Bayesian Information Criterition
```
